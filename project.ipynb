```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('train.csv')
```

```python
data
```

```python
# 1. Display basic information about the dataset
print(data.info())
```

```python
# 2. Display the first few rows of the dataset
print(data.head())
```

```python
# 3. Check for missing values
print(data.isnull().sum())
```

```python
# 4. Calculate summary statistics
print(data.describe())
```

```python
# 5. Count unique values in a column
print(data['category_id'].value_counts())
```

```python
# 6. Average views per category
data['views'] = pd.to_numeric(data['views'], errors='coerce')
category_views = data.groupby('category_id')['views'].mean()
print(category_views)
```

```python
# 7. Correlation between views and likes
data['views'] = pd.to_numeric(data['views'], errors='coerce')
data['likes'] = pd.to_numeric(data['likes'], errors='coerce')
correlation = data['views'].corr(data['likes'])
print(correlation)
```

```python
#8. Videos with the highest likes
most_liked_videos = data.nlargest(10, 'likes')
print(most_liked_videos[['title', 'likes']])
```

```python
# 9. Videos with the most comments
data['comment_count'] = pd.to_numeric(data['comment_count'], errors='coerce')
most_commented_videos = data.nlargest(10, 'comment_count')
print(most_commented_videos[['title', 'comment_count']])
```

```python
#10 Create a correlation matrix heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 11. Create a bar plot of average views per category
category_views = data.groupby('category_id')['views'].mean()
plt.figure(figsize=(10, 6))
category_views.plot(kind='bar')
plt.title('Average Views per Category')
plt.ylabel('Average Views')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#12. Create a bar plot of video count by trend day count
trend_day_counts = data['Trend_day_count'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
trend_day_counts.plot(kind='bar')
plt.title('Video Count by Trend Day Count')
plt.xlabel('Trend Day Count')
plt.ylabel('Video Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#13 Create a bar plot of video count per category
category_counts = data['category_id'].value_counts()
plt.figure(figsize=(12, 6))
category_counts.plot(kind='bar')
plt.title('Video Count per Category')
plt.xlabel('Category')
plt.ylabel('Video Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#14 Create a scatter plot of subscribers vs. views
data['subscriber'] = pd.to_numeric(data['subscriber'], errors='coerce')
data['views'] = pd.to_numeric(data['views'], errors='coerce')

# Create a scatter plot with larger dots and hue
plt.figure(figsize=(10,10))
sns.scatterplot(data=data, x='subscriber', y='views', hue='category_id', s=100, alpha=0.5)
plt.title('Subscribers vs. Views')
plt.xlabel('Subscribers')
plt.ylabel('Views')
plt.legend(title='Category')
plt.show()
```

```python

data['likes'] = pd.to_numeric(data['likes'], errors='coerce')
data['dislike'] = pd.to_numeric(data['dislike'], errors='coerce')

# Calculate average likes and dislikes per category
avg_likes_per_category = data.groupby('category_id')['likes'].mean()
avg_dislikes_per_category = data.groupby('category_id')['dislike'].mean()

# Create a DataFrame for bar plot
bar_plot_data = pd.DataFrame({
    'Average Likes': avg_likes_per_category,
    'Average Dislikes': avg_dislikes_per_category
})

# Plot the bar plot
plt.figure(figsize=(10, 6))
bar_plot_data.plot(kind='bar')
plt.title('Average Likes and Dislikes per Category')
plt.ylabel('Count')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
#15. Count of treding days of videos
data['Trend_day_count'] = pd.to_numeric(data['Trend_day_count'], errors='coerce')
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Trend_day_count')
plt.title('Count Plot of Trending Days')
plt.xlabel('Trending Days')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
